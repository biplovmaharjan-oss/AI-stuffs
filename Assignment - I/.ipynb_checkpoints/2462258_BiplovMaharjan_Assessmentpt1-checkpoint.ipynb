```python
# Identify columns to impute (those with missing values)
# List of columns to impute missing values
cols_to_impute = [
    'hdi', 'hdi_f', 'hdi_m', 'expec_yr_school', 'expec_yr_school_f',
    'expec_yr_school_m', 'mean_yr_school', 'mean_yr_school_f',
    'mean_yr_school_m', 'gross_inc_percap', 'gross_inc_percap_f',
    'gross_inc_percap_m', 'gender_development', 'gender_inequality',
    'secondary_education_f_%', 'secondary_education_m_%',
    'seats_in_parliament_f_%', 'seats_in_parliament_m_%',
    'labour_participation_f_%', 'labour_participation_m_%',
    'co2_emission_tons', 'mat_footprint_percap_tons'
]

# Store initial number of rows
initial_rows = hdi_2022_df.shape[0]

# Drop aggregate rows (ISO codes starting with 'ZZ')
mask_to_drop = hdi_2022_df['iso3'].str.startswith('ZZ', na=False)
print(f"Rows to drop (aggregates): {mask_to_drop.sum()}")
hdi_2022_df = hdi_2022_df.loc[~mask_to_drop].copy()
print(f"Rows after removing aggregates: {hdi_2022_df.shape[0]}")

# Impute missing values with median for each column (avoid chained assignment)
for col in cols_to_impute:
    median_val = hdi_2022_df[col].median()
    hdi_2022_df[col] = hdi_2022_df[col].fillna(median_val)  # assign back to column
    print(f"     - Imputed '{col}' with median: {median_val:.4f}")

# Count rows removed and check missing values
rows_removed = initial_rows - hdi_2022_df.shape[0]
print(f"\nRows removed due to missing critical data (country, hdi, income): {rows_removed}")
print("Final missing value count per column:")
print(hdi_2022_df.isnull().sum())

# Display final shape
print(f"\n3. FINAL SHAPE AFTER CLEANING: {hdi_2022_df.shape}\nInitial shape was {initial_rows}")

# Save cleaned dataset
hdi_2022_df.to_csv("cleaned.csv", index=False)

```

## Task:6 - Adding HDI Category Column

```python
def hdi_category(hdi):
    """
    Categorize a country based on its Human Development Index (HDI) value.

    Parameters:
    -----------
    hdi : float
        The HDI value of a country.

    Returns:
    --------
    str
        The HDI category as a string:
        - "Low" for HDI < 0.550
        - "Medium" for 0.550 <= HDI < 0.700
        - "High" for 0.700 <= HDI < 0.800
        - "Very High" for HDI >= 0.800
    """
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

# Apply HDI categorization to the dataset
hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)

# Verification of category counts
hdi_2022_df['HDI Category'].value_counts()
```

```python
# Save final CSV
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
print("Dataset with HDI Category column is saved as CSV")
```

## Task:3 - Outlier Detection

```python
#Detecting outliers in HDI and GNI per Capita
# Function to detect outliers using the IQR method
def detect_outliers(series):
    """
    Identify outliers in a numerical pandas Series using the IQR method.

    Parameters:
    -----------
    series : pandas.Series
        The numeric data for which outliers are to be detected.

    Returns:
    --------
    pandas.Series (bool)
        A boolean Series where True indicates that the value is an outlier
        (below Q1 - 1.5*IQR or above Q3 + 1.5*IQR), and False otherwise.
    
    Notes:
    ------
    The interquartile range (IQR) is calculated as Q3 - Q1.
    Any value outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR] is considered an outlier.
    """
    # Calculate first and third quartiles
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    
    # Compute interquartile range
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Return a boolean series indicating outliers
    return (series < lower) | (series > upper)


# Detect HDI outliers
southasia_latest['HDI_outlier'] = detect_outliers(southasia_latest['hdi'])

# Detect GNI per capita outliers
southasia_latest['GNI_outlier'] = detect_outliers(southasia_latest['gross_inc_percap'])
```

```python
#Creating scatter plot of GNI per Capita vs HDI
# Create figure
plt.figure(figsize=(8,6))

# Scatter plot of HDI vs GNI per capita
sns.scatterplot(
    data=southasia_latest,
    x='gross_inc_percap',
    y='hdi',
    hue='HDI_outlier',
    palette=['blue']
)

# Add title and labels
plt.title("Outliers in HDI vs GNI per Capita (South Asia, 2022)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")

# Adjust layout and show plot
plt.tight_layout()
plt.savefig('Outlier_in_HDI_vs_GNI_per_capita.png', dpi=300)
plt.show()
```

## Task:5 - HDI disparity

```python
# Computing HDI ranges
# Function to calculate HDI range (max - min)
# Function to calculate HDI range
def hdi_range(df):
    max_hdi = df['hdi'].max()   # Highest HDI
    min_hdi = df['hdi'].min()   # Lowest HDI
    return max_hdi - min_hdi    # Range

# Calculate HDI range for each region
southasia_range = hdi_range(southasia_2020_2022)
middleeast_range = hdi_range(middleeast_2020_2022)

# Print results
print("South Asia HDI Range:", southasia_range)
print("Middle East HDI Range:", middleeast_range)

```

```python
#Computing Coefficient
# Function to calculate HDI coefficient of variation
def hdi_cv(df):
    """
    Calculate the coefficient of variation (CV) for the HDI in a given DataFrame.

    Parameters:
    -----------
    df : pandas.DataFrame
        A DataFrame containing a column named 'hdi' with Human Development Index values.

    Returns:
    --------
    float
        The coefficient of variation of HDI, computed as the standard deviation
        divided by the mean (CV = std / mean). This measures relative variability
        of HDI values within the dataset.
    """
    hdi_std = df['hdi'].std()    # Standard deviation of HDI
    hdi_mean = df['hdi'].mean()  # Mean of HDI
    return hdi_std / hdi_mean    # CV = std / mean

# Calculate HDI CV for each region
southasia_cv = hdi_cv(southasia_2020_2022)
middleeast_cv = hdi_cv(middleeast_2020_2022)

# Print results
print("South Asia HDI Coefficient of Variation (CV):", southasia_cv)
print("Middle East HDI Coefficient of Variation (CV):", middleeast_cv)

```


-HDI varies more in the Middle East than in South Asia.

## Task:6 - Correlation Analysis

```python
#Computing correlations of HDI with gender development anf life expectancy
# Function to perform correlation analysis between HDI and selected metrics
def correlation_analysis(df, region_name):
    metrics = ['gender_development', 'life_expectancy']  # Metrics to analyze
    
    for metric in metrics:
        corr = df[metric].corr(df['hdi'])                # Compute correlation
        print(f"{region_name} | HDI vs {metric}: {corr:.3f}")

# Perform correlation analysis for each region
correlation_analysis(southasia_2020_2022, "South Asia")
correlation_analysis(middleeast_2020_2022, "Middle East")
```

## Task:7 - Outliers Detection

```python
#detectiong outliers in hdi and GNI
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    return (series < Q1 - 1.5*IQR) | (series > Q3 + 1.5*IQR)
```

```python
#Creating scatter plot highlighting outliers
# Plot HDI vs GNI and highlight outliers for each region
regions = [
    (southasia_2020_2022, "South Asia"),
    (middleeast_2020_2022, "Middle East")
]

n_rows = len(regions)
fig, axes = plt.subplots(n_rows, 1, figsize=(8, 6*n_rows))  # 1 column, multiple rows

for i, (region_df, name) in enumerate(regions):
    # Detect outliers
    region_df['Outlier'] = (
        detect_outliers(region_df['hdi']) |
        detect_outliers(region_df['gross_inc_percap'])
    )
    
    # Dynamic palette
    unique_values = region_df['Outlier'].unique()
    palette = ['blue', 'red'][:len(unique_values)]
    
    ax = axes[i] if n_rows > 1 else axes
    sns.scatterplot(
        data=region_df,
        x='gross_inc_percap',
        y='hdi',
        hue='Outlier',
        palette=palette,
        ax=ax
    )
    
    ax.set_title(f"HDI vs GNI Outliers ({name})")
    ax.legend(title='Outlier')

plt.tight_layout()
plt.savefig('hdi_gni_outliers_combined.png', dpi=300)
plt.show()
```
