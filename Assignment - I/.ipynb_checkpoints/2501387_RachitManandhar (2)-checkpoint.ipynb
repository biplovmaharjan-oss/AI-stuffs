## 3.1 Problem 1-A: Single Year HDI Exploration (Latest Year: 2022):

### Task 1: Extract Latest Year:

```python
print(sorted(df["year"].unique()))
```

```python
hdi_2022_df = df[df["year"] == 2022]
hdi_2022_df = hdi_2022_df.copy()
```

### Task 2: Data Exploration:

```python
hdi_2022_df.head(10)
```

```python
rows, columns = hdi_2022_df.shape
print("Number of rows: ", rows)
print("Number of columns: ", columns)
```

```python
hdi_2022_df.info()
```

### Task 3: Missing values & Data Cleaning:

```python
print(hdi_2022_df.isna().sum())
```

```python
# 1. Replace special missing-value characters
hdi_2022_df.replace("-", pd.NA, inplace=True)

# 2. Select numeric columns (exclude identifiers)
numerical_columns = hdi_2022_df.columns.drop(["iso3", "country"])

# 3. Convert to numeric
hdi_2022_df[numerical_columns] = hdi_2022_df[numerical_columns].apply(pd.to_numeric, errors="coerce")

# 4. Clean country names
hdi_2022_df["country"] = hdi_2022_df["country"].str.strip()

# 5. Drop rows missing key identifiers or core indicator
hdi_2022_df.dropna(subset=["hdi", "iso3", "country"], inplace=True)

# 6. Remove duplicate rows
hdi_2022_df.drop_duplicates(inplace=True)

# 7. Median imputation for remaining numeric columns
hdi_2022_df[numerical_columns] = hdi_2022_df[numerical_columns].fillna(hdi_2022_df[numerical_columns].median())
```

### Task 4: Basic Statistics:

```python
hdi_2022_mean = hdi_2022_df["hdi"].mean()
hdi_2022_median = hdi_2022_df["hdi"].median()
hdi_2022_std = hdi_2022_df["hdi"].std()

print(f"Mean HDI: {hdi_2022_mean:.3f} \nMedian HDI: {hdi_2022_median:.3f} \nStandard Deviation: {hdi_2022_std:.3f}")

hdi_2022_highest = hdi_2022_df.loc[hdi_2022_df["hdi"].idxmax()]
hdi_2022_lowest = hdi_2022_df.loc[hdi_2022_df["hdi"].idxmin()]

print(f"Country with Highest HDI is: {hdi_2022_highest["country"]} ({hdi_2022_highest["hdi"]:.3f})")
print(f"Country with Lowest HDI is: {hdi_2022_lowest["country"]} ({hdi_2022_lowest["hdi"]:.3f})")
```

### Task 5: Filtering and Sorting:

```python
hdi_2022_high = hdi_2022_df[hdi_2022_df["hdi"] > 0.800].copy()
hdi_2022_high = hdi_2022_high.sort_values("gross_inc_percap", ascending = False)
print(hdi_2022_high[["country", "hdi", "gross_inc_percap"]].head(10))
```

### Task 6: Adding HDI Category Column:

```python
def categorize(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi >= 0.550 and hdi < 0.700:
        return "Medium"
    elif hdi >= 0.700 and hdi < 0.800:
        return "High"
    else: 
        return "Very High"

hdi_2022_df["HDI_Category"] = hdi_2022_df["hdi"].apply(categorize)

print(hdi_2022_df[["hdi", "HDI_Category"]])
```

### Task 4: Short Analysis Questions:

```python
print("1. Countries with greatest HDI improvement 2020â€“2022:")

hdi_changes = {}

for country in HDI_Problem1B['country'].unique():
    country_data = HDI_Problem1B[HDI_Problem1B['country'] == country]
    hdi_2020 = country_data[country_data['year'] == 2020]['hdi']
    hdi_2022 = country_data[country_data['year'] == 2022]['hdi']

    if len(hdi_2020) > 0 and len(hdi_2022) > 0:
        change = hdi_2022.values[0] - hdi_2020.values[0]
        hdi_changes[country] = change

# Sort and show top 5 improvers
sorted_improvement = sorted(hdi_changes.items(), key=lambda x: x[1], reverse=True)
for i, (country, change) in enumerate(sorted_improvement[:5], 1):
    print(f"\t{i}. {country}: +{change:.4f}")
```

```python
print("\n2. Countries with HDI decline 2020â€“2022:")

decliners = [(country, change) for country, change in hdi_changes.items() if change < 0]

if decliners:
    # Show up to 5 biggest declines
    for i, (country, change) in enumerate(sorted(decliners, key=lambda x: x[1])[:5], 1):
        print(f"\t{i}. {country}: {change:.4f}")
else:
    print("\tNo countries showed decline in this period.")
```

<p>These countries could have experienced a decline in HDI due to war, political instability, economic crises, and COVID-19 impacts. These drops reflect setbacks in health, education, and income.</p>

```python
print("\n3. Regional HDI analysis (2020â€“2022):")

# Average HDI by Region and Year
hdi_region_year_avg = (
    HDI_Problem1B
    .groupby(['region', 'year'])['hdi']
    .mean()
    .reset_index()
)

# Average across 2020â€“2022 for each Region
hdi_region_avg = hdi_region_year_avg.groupby('region')['hdi'].mean()

highest_region = hdi_region_avg.idxmax()
lowest_region  = hdi_region_avg.idxmin()

print("\tAverage HDI by region across 2020â€“2022:")
for region, val in hdi_region_avg.items():
    print(f"\t-{region}: {val:.3f}")

print(f"\n\tHighest average HDI: {highest_region} ({hdi_region_avg[highest_region]:.3f})")
print(f"\tLowest average HDI : {lowest_region} ({hdi_region_avg[lowest_region]:.3f})")
```

Impact of Covid-19 Pandemic on HDI:
<p>The Covid-19 pandemic may have affected HDI trends due to higher death rates, halt to educational institutions, lower income per capita and loss of jobs.</p>

```python
# Compare the ranking of countries by Composite Score with their HDI ranking and discuss any differences.

# Compute ranks
south_asia_ranked["composite_rank"] = south_asia_ranked["Composite_Score"].rank(ascending=False, method="min").astype(int)
south_asia_ranked["hdi_rank"] = south_asia_ranked["hdi"].rank(ascending=False, method="min").astype(int)

# Sort by Composite Score rank (optional)
south_asia_ranked = south_asia_ranked.sort_values("composite_rank").reset_index(drop=True)

south_asia_ranked[["country", "Composite_Score", "hdi", "composite_rank", "hdi_rank"]]
```

<b>Discussion:</b> There maybe differences between hdi and composite score rankings as composite score only considers life expectancy and gross income per capita, so it can over or under rank where education matters.

## Task 3: Outlier Detection:

```python
# Create a scatter plot of GNI per Capita vs HDI, highlighting the outliers in a different color.

plt.figure(figsize=(10, 6))

south_asia_latest_clean = south_asia_latest.dropna(
    subset=['hdi', 'gross_inc_percap']
)

# IQR bounds for HDI
Q1_hdi = south_asia_latest_clean['hdi'].quantile(0.25)
Q3_hdi = south_asia_latest_clean['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi
hdi_bounds = (Q1_hdi - 1.5 * IQR_hdi, Q3_hdi + 1.5 * IQR_hdi)

# IQR bounds for GNI per capita
Q1_gni = south_asia_latest_clean['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_latest_clean['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
gni_bounds = (Q1_gni - 1.5 * IQR_gni, Q3_gni + 1.5 * IQR_gni)

# Outlier mask
outlier_mask = (
    (south_asia_latest_clean['hdi'] < hdi_bounds[0]) |
    (south_asia_latest_clean['hdi'] > hdi_bounds[1]) |
    (south_asia_latest_clean['gross_inc_percap'] < gni_bounds[0]) |
    (south_asia_latest_clean['gross_inc_percap'] > gni_bounds[1])
)

# Normal points
plt.scatter(
    south_asia_latest_clean[~outlier_mask]['gross_inc_percap'],
    south_asia_latest_clean[~outlier_mask]['hdi'],
    alpha=0.7,
    label='Normal',
    color='blue'
)

# Outliers
plt.scatter(
    south_asia_latest_clean[outlier_mask]['gross_inc_percap'],
    south_asia_latest_clean[outlier_mask]['hdi'],
    alpha=0.9,
    label='Outliers',
    color='red',
    s=100
)

# Label outlier countries
for _, row in south_asia_latest_clean[outlier_mask].iterrows():
    plt.annotate(
        row['country'],
        (row['gross_inc_percap'], row['hdi']),
        xytext=(5, 5),
        textcoords='offset points'
    )

plt.xlabel('GNI per Capita (USD)')
plt.ylabel('Human Development Index (HDI)')
plt.title('GNI per Capita vs HDI with Outliers Highlighted (South Asia)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

<b>Discussion:</b> The Identified countries stand out as outliers as their GNI per capita or HDI falls outside the 1.5 X IQR .They are different from regional cluster which reflects extreme economic and development conditions caused by many factors.

## Task 4: Exploring Metric Relationships:

```python
# Use latest year and drop missing values
correlation_data = south_asia_latest.dropna(subset=['hdi', 'gender_development', 'life_expectancy'])

corr_gender_hdi = correlation_data['gender_development'].corr(correlation_data['hdi'])
corr_life_hdi = correlation_data['life_expectancy'].corr(correlation_data['hdi'])

print(f"Correlation between Gender Development Index and HDI: {corr_gender_hdi:.3f}")
print(f"Correlation between Life Expectancy and HDI: {corr_life_hdi:.3f}")
```

```python
# Create scatter plots with trendlines to visualize the relationships
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Gender Development vs HDI
axes[0].scatter(correlation_data['gender_development'], correlation_data['hdi'], alpha=0.7)
z_gender = np.polyfit(correlation_data['gender_development'], correlation_data['hdi'], 1)
p_gender = np.poly1d(z_gender)
axes[0].plot(
    correlation_data['gender_development'],
    p_gender(correlation_data['gender_development']),
    "r--",
    alpha=0.8
)
axes[0].set_xlabel('Gender Development Index')
axes[0].set_ylabel('HDI')
axes[0].set_title(f'Gender Development vs HDI (r={corr_gender_hdi:.3f})')
axes[0].grid(True, alpha=0.3)

# Plot 2: Life Expectancy vs HDI
axes[1].scatter(correlation_data['life_expectancy'], correlation_data['hdi'], alpha=0.7)
z_life = np.polyfit(correlation_data['life_expectancy'], correlation_data['hdi'], 1)
p_life = np.poly1d(z_life)
axes[1].plot(
    correlation_data['life_expectancy'],
    p_life(correlation_data['life_expectancy']),
    "r--",
    alpha=0.8
)
axes[1].set_xlabel('Life Expectancy')
axes[1].set_ylabel('HDI')
axes[1].set_title(f'Life Expectancy vs HDI (r={corr_life_hdi:.3f})')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('metric_relationships.png', dpi=300)
plt.show()
```

<b>Discussion:</b> Both Life Expectancy and Gender Development Index show a strong positive correlatio with HDI, but Life Expectancy is slightly more related.

## Task 2: Descriptive Statistics:

## Task 3: Top and Bottom Performers:

```python
#  Identify the top 3 and bottom 3 countries in each region based on HDI

def get_top_bottom_performers(region_df, n=3):

    """
    Compute average HDI (2020â€“2022) per country and
    return the top n and bottom n countries.
    """
    country_avg_hdi = (
        region_df.groupby("country")["hdi"]
        .mean()
        .sort_values(ascending=False)
    )
    top_countries = country_avg_hdi.head(n)
    bottom_countries = country_avg_hdi.tail(n)
    return top_countries, bottom_countries


south_asia_top, south_asia_bottom = get_top_bottom_performers(
    HDI_SouthAsia_2020_2022
)
print("South Asia â€“ Top 3 performers:")
print(south_asia_top)
print("\nSouth Asia â€“ Bottom 3 performers:")
print(south_asia_bottom)

middle_east_top, middle_east_bottom = get_top_bottom_performers(
    HDI_MiddleEast_2020_2022
)
print("\nMiddle East â€“ Top 3 performers:")
print(middle_east_top)
print("\nMiddle East â€“ Bottom 3 performers:")
print(middle_east_bottom)
```

## Task 5: HDI Disparity:

```python
#Compute the range (maxâ€“ min) of HDI for each region.
#Compute the coefficient of variation (CV = std/mean) for HDI

def calculate_hdi_disparity(region_df, region_name):
    """
    Compute HDI range, mean, standard deviation, and
    coefficient of variation for a given region.
    """
    hdi_values = region_df["hdi"].dropna()

    hdi_range = hdi_values.max() - hdi_values.min()
    hdi_mean = hdi_values.mean()
    hdi_std = hdi_values.std()
    coefficient_variation = (hdi_std / hdi_mean) * 100 if hdi_mean != 0 else 0

    return {
        "Region": region_name,
        "Range": hdi_range,
        "Mean": hdi_mean,
        "Std": hdi_std,
        "CV (%)": coefficient_variation,
    }

# Compute HDI disparity stats for both regions
south_asia_stats = calculate_hdi_disparity(HDI_SouthAsia_2020_2022, "South Asia")
middle_east_stats = calculate_hdi_disparity(HDI_MiddleEast_2020_2022, "Middle East")

print("HDI disparity analysis:")
print(f"{'Statistic':15} {'South Asia':>12} {'Middle East':>12}")
print("-" * 45)

for key in ["Mean", "Range", "Std", "CV (%)"]:
    print(f"{key:15} {south_asia_stats[key]:12.3f} {middle_east_stats[key]:12.3f}")
```

```python
# Identify which region exhibits more variation in HDI
if south_asia_stats["CV (%)"] > middle_east_stats["CV (%)"]:
    print(
        f"\nSouth Asia exhibits more variation in HDI (CV: {south_asia_stats['CV (%)']:.1f}% vs {middle_east_stats['CV (%)']:.1f}%)"
    )
else:
    print(
        f"\nMiddle East exhibits more variation in HDI (CV: {middle_east_stats['CV (%)']:.1f}% vs {south_asia_stats['CV (%)']:.1f}%)"
    )
```

## Task 6: Correlation Analysis:

```python
# For each region, compute correlations of HDI with:â€“ Gender Development Indexâ€“ Life Expectancy Index

def calculate_correlations(region_df):
    """
    Compute correlations of HDI with gender_development and life_expectancy
    for a given regional subset.
    """
    corr_results = {}

    # Gender Development vs HDI
    if "gender_development" in region_df.columns:
        region_df["gender_development"] = pd.to_numeric(region_df["gender_development"], errors="coerce")
        gd_data = region_df[["gender_development", "hdi"]].dropna()
        corr_results["Gender Development vs HDI"] = gd_data.corr().iloc[0, 1]

    # Life Expectancy vs HDI
    region_df["life_expectancy"] = pd.to_numeric(region_df["life_expectancy"], errors="coerce")
    le_data = region_df[["life_expectancy", "hdi"]].dropna()
    corr_results["Life Expectancy vs HDI"] = le_data.corr().iloc[0, 1]

    return corr_results

# Compute correlations for each region
south_asia_corr = calculate_correlations(HDI_SouthAsia_2020_2022)
middle_east_corr = calculate_correlations(HDI_MiddleEast_2020_2022)

# Print correlation table
print("Correlation Analysis:")
print(f"{'Correlation':30} {'South Asia':>12} {'Middle East':>12}")
print("-" * 60)

for metric in ["Gender Development vs HDI", "Life Expectancy vs HDI"]:
    sa_val = south_asia_corr.get(metric, float("nan"))
    me_val = middle_east_corr.get(metric, float("nan"))
    print(f"{metric:30} {sa_val:12.3f} {me_val:12.3f}")
```

```python
def scatter_with_trend(ax, df, x, y, title):
    data = df[[x, y]].dropna()
    ax.scatter(data[x], data[y], alpha=0.6)
    
    z = np.polyfit(data[x], data[y], 1)
    p = np.poly1d(z)
    ax.plot(data[x], p(data[x]), "r--", alpha=0.8)
    
    ax.set_xlabel(x.replace("_", " ").title())
    ax.set_ylabel("HDI")
    ax.set_title(title)
    ax.grid(True, alpha=0.3)


fig, axes = plt.subplots(2, 2, figsize=(12, 10))

scatter_with_trend(
    axes[0, 0],
    HDI_SouthAsia_2020_2022,
    "gender_development",
    "hdi",
    f"South Asia: Gender Dev vs HDI (r={south_asia_corr['Gender Development vs HDI']:.3f})"
)

scatter_with_trend(
    axes[0, 1],
    HDI_SouthAsia_2020_2022,
    "life_expectancy",
    "hdi",
    f"South Asia: Life Expectancy vs HDI (r={south_asia_corr['Life Expectancy vs HDI']:.3f})"
)

scatter_with_trend(
    axes[1, 0],
    HDI_MiddleEast_2020_2022,
    "gender_development",
    "hdi",
    f"Middle East: Gender Dev vs HDI (r={middle_east_corr['Gender Development vs HDI']:.3f})"
)

scatter_with_trend(
    axes[1, 1],
    HDI_MiddleEast_2020_2022,
    "life_expectancy",
    "hdi",
    f"Middle East: Life Expectancy vs HDI (r={middle_east_corr['Life Expectancy vs HDI']:.3f})"
)

plt.tight_layout()
plt.savefig("correlation_analysis.png", dpi=300)
plt.show()
```

<b>Interpretation:</b> As shown by the high correlation coefficients (r = 0.874 - 0.939), the relationships are strong and positive. It shows that higher gender development and life expectancy are associated with higher HDI.

## Task 7: Outlier Detection:

```python
def iqr_bounds(series):
    Q1, Q3 = series.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    return Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
```

```python
def plot_region_outliers(ax, df, region_name, color):
    data = df.dropna(subset=["hdi", "gross_inc_percap"]).copy()

    hdi_low, hdi_high = iqr_bounds(data["hdi"])
    gni_low, gni_high = iqr_bounds(data["gross_inc_percap"])

    outliers = (
        (data["hdi"] < hdi_low) | (data["hdi"] > hdi_high) |
        (data["gross_inc_percap"] < gni_low) | (data["gross_inc_percap"] > gni_high)
    )

    ax.scatter(
        data.loc[~outliers, "gross_inc_percap"],
        data.loc[~outliers, "hdi"],
        alpha=0.7,
        label="Normal",
        color=color
    )

    ax.scatter(
        data.loc[outliers, "gross_inc_percap"],
        data.loc[outliers, "hdi"],
        alpha=0.9,
        label="Outliers",
        color="red",
        s=100
    )

    ax.set_title(f"{region_name}: GNI per Capita vs HDI")
    ax.set_xlabel("GNI per Capita (USD)")
    ax.set_ylabel("HDI")
    ax.grid(True, alpha=0.3)
    ax.legend()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

plot_region_outliers(
    axes[0],
    HDI_SouthAsia_2020_2022,
    "South Asia",
    color="blue"
)

plot_region_outliers(
    axes[1],
    HDI_MiddleEast_2020_2022,
    "Middle East",
    color="green"
)

plt.tight_layout()
plt.show()
```

<b>Discussion:</b> The outlier analysis shows that South Asia has no extreme outliers in neither HDI nor GNI per capita suggesting relatively consistent development levels across countries in the region. However, in Middle East Yemen shows an outlier with values far below its region's range. These outliers show the development conditions of the country which differs significantly from their regional norms where economic wealth does not fully translate into Human Development, like health and education.
