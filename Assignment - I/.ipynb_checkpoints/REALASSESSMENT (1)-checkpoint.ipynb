# Problem 1: Basic Data Exploration & Trend Visualization

## Problem - 1A - Single Year HDI Exploration (Latest Year: 2022)
Objective: Explore the HDI dataset for the latest available year (2022) to practice basic EDA techniques.

### Task 1: Extract latest year (2022)

```python
import pandas as pd
df = pd.read_csv('Human_Development_Index_Dataset.csv', encoding='latin1')

# Identify unique years
unique_years = sorted(df['year'].unique())
print("Unique years in dataset:", unique_years)
print("Latest year:", max(unique_years))

# Filter for 2022
hdi_2022_df = df[df['year'] == 2022].copy()
print("Countries in 2022:", len(hdi_2022_df))
```

### Task2: Data Exploration

```python
# Display first 10 rows
print("First 10 rows of 2022 dataset:")
print(hdi_2022_df.head(10))

# Count rows and columns
print(f"\nRows: {hdi_2022_df.shape[0]}, Columns: {hdi_2022_df.shape[1]}")

# List column names and data types
print("\nColumn names and data types:")
for col in hdi_2022_df.columns:
    print(f"{col}: {hdi_2022_df[col].dtype}")
```

### Task 3: Missing Values & Data Cleaning

```python
# Check for missing values
print("Missing values in each column:")
missing_values = hdi_2022_df.isnull().sum()
print(missing_values[missing_values > 0])

# Inspect dataset for issues
print(f"\nDuplicate rows: {hdi_2022_df.duplicated().sum()}")

# Convert numeric columns stored as text
numeric_cols = ['hdi', 'life_expectancy', 'pop_millions', 'gross_inc_percap']
for col in numeric_cols:
    if col in hdi_2022_df.columns:
        hdi_2022_df[col] = pd.to_numeric(hdi_2022_df[col], errors='coerce')

# Remove rows with missing HDI
rows_before = len(hdi_2022_df)
hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])
rows_after = len(hdi_2022_df)
print(f"\nDropped {rows_before - rows_after} rows with missing HDI values")

# Remove duplicates
hdi_2022_df = hdi_2022_df.drop_duplicates()
print(f"Final rows after cleaning: {len(hdi_2022_df)}")
```

### Task 4: Basic Statistics

```python
# Compute mean, median, and standard deviation of HDI
mean_hdi = hdi_2022_df['hdi'].mean()
median_hdi = hdi_2022_df['hdi'].median()
std_hdi = hdi_2022_df['hdi'].std()

print(f"Mean HDI: {mean_hdi:.3f}")
print(f"Median HDI: {median_hdi:.3f}")
print(f"Standard Deviation: {std_hdi:.3f}")

# Country with highest HDI
highest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()]
print(f"\nHighest HDI: {highest_hdi_country['country']} ({highest_hdi_country['hdi']:.3f})")

# Country with lowest HDI
lowest_hdi_country = hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()]
print(f"Lowest HDI: {lowest_hdi_country['country']} ({lowest_hdi_country['hdi']:.3f})")
```

### Task 5: Filtering and Sorting

```python
# Filter countries with HDI > 0.800
high_hdi_countries = hdi_2022_df[hdi_2022_df['hdi'] > 0.800].copy()
print(f"Countries with HDI > 0.800: {len(high_hdi_countries)}")

# Sort by Gross Income per Capita descending
high_hdi_sorted = high_hdi_countries.sort_values('gross_inc_percap', ascending=False)

# Display top 10 countries
print("\nTop 10 countries with HDI > 0.800 (sorted by GNI per capita):")
print(high_hdi_sorted[['country', 'hdi', 'gross_inc_percap']].head(10))
```

### Task 6: Adding HDI Category Column

```python
# Function to categorize HDI
def categorize_hdi(hdi_value):
    if pd.isna(hdi_value):
        return 'Unknown'
    elif hdi_value < 0.550:
        return 'Low'
    elif 0.550 <= hdi_value < 0.699:
        return 'Medium'
    elif 0.700 <= hdi_value < 0.799:
        return 'High'
    else:
        return 'Very High'

# Add HDI Category column
hdi_2022_df['HDI_Category'] = hdi_2022_df['hdi'].apply(categorize_hdi)

# Verify classification
print("HDI Category Distribution:")
print(hdi_2022_df['HDI_Category'].value_counts())

# Save final dataframe
hdi_2022_df.to_csv('HDI_category_added.csv', index=False)
print("\nSaved to 'HDI_category_added.csv'")
```

## TASK 2: Composite Development Score

```python

# Task 2.1: Create Composite Score
# Normalize life expectancy (scale 0-1, assuming typical range 50-85)
hdi_south_asia['Life_Expectancy_Index'] = (hdi_south_asia['life_expectancy'] - 50) / (85 - 50)
hdi_south_asia['Life_Expectancy_Index'] = hdi_south_asia['Life_Expectancy_Index'].clip(0, 1)

# Normalize GNI per capita (scale 0-1, assuming typical range 1000-50000)
hdi_south_asia['GNI_Index'] = (hdi_south_asia['gross_inc_percap'] - 1000) / (50000 - 1000)
hdi_south_asia['GNI_Index'] = hdi_south_asia['GNI_Index'].clip(0, 1)

# Calculate Composite Score
hdi_south_asia['Composite_Score'] = (0.30 * hdi_south_asia['Life_Expectancy_Index'] +
                                     0.30 * hdi_south_asia['GNI_Index'])

# Task 2.2: Rank South Asian countries based on Composite Score (using latest year)
latest_year = hdi_south_asia['year'].max()
south_asia_latest = hdi_south_asia[hdi_south_asia['year'] == latest_year].copy()

# Rank by Composite Score
south_asia_ranked = south_asia_latest.sort_values('Composite_Score', ascending=False)
print("\nSouth Asian countries ranked by Composite Score:")
print(south_asia_ranked[['country', 'Composite_Score', 'hdi']].to_string(index=False))

# Task 2.3: Plot top 5 countries in horizontal bar chart
top_5 = south_asia_ranked.head(5)

plt.figure(figsize=(10, 6))
plt.barh(top_5['country'], top_5['Composite_Score'], color='skyblue')
plt.xlabel('Composite Score')
plt.title('Top 5 South Asian Countries by Composite Score')
plt.gca().invert_yaxis()  # Highest score at top
plt.tight_layout()
plt.savefig('composite_score_top5.png', dpi=300)
plt.show()

# Task 2.4: Compare Composite Score ranking with HDI ranking
south_asia_ranked_hdi = south_asia_latest.sort_values('hdi', ascending=False)
print("\nComparison of rankings:")
print("Country | Composite Rank | HDI Rank | Composite Score | HDI")
print("-" * 60)

composite_ranks = {country: rank+1 for rank, country in enumerate(south_asia_ranked['country'])}
hdi_ranks = {country: rank+1 for rank, country in enumerate(south_asia_ranked_hdi['country'])}

for country in south_asia_ranked['country']:
    comp_score = south_asia_ranked[south_asia_ranked['country'] == country]['Composite_Score'].values[0]
    hdi_value = south_asia_ranked[south_asia_ranked['country'] == country]['hdi'].values[0]
    print(f"{country:15} | {composite_ranks[country]:14} | {hdi_ranks[country]:8} | {comp_score:.3f}          | {hdi_value:.3f}")

print("\nDiscussion: Countries may rank differently because Composite Score weights life expectancy")
print("and GNI equally at 30% each, while HDI includes education components. Countries with")
print("high life expectancy and income but lower education may rank higher in Composite Score.")
```

## TASK 3: Outlier Detection


```python
# Task 3.1: Detect outliers using 1.5 Ã— IQR rule
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers in HDI
hdi_outliers = detect_outliers_iqr(hdi_south_asia, 'hdi')
print("Outliers in HDI:")
print(hdi_outliers[['country', 'year', 'hdi']].to_string(index=False))

# Detect outliers in GNI per Capita
gni_outliers = detect_outliers_iqr(hdi_south_asia, 'gross_inc_percap')
print("\nOutliers in GNI per Capita:")
print(gni_outliers[['country', 'year', 'gross_inc_percap']].to_string(index=False))

# Task 3.2: Create scatter plot highlighting outliers
plt.figure(figsize=(10, 6))

# Use latest year data
south_asia_latest_clean = south_asia_latest.dropna(subset=['hdi', 'gross_inc_percap'])

# Identify outliers in the scatter plot
Q1_hdi = south_asia_latest_clean['hdi'].quantile(0.25)
Q3_hdi = south_asia_latest_clean['hdi'].quantile(0.75)
IQR_hdi = Q3_hdi - Q1_hdi
hdi_bounds = (Q1_hdi - 1.5 * IQR_hdi, Q3_hdi + 1.5 * IQR_hdi)

Q1_gni = south_asia_latest_clean['gross_inc_percap'].quantile(0.25)
Q3_gni = south_asia_latest_clean['gross_inc_percap'].quantile(0.75)
IQR_gni = Q3_gni - Q1_gni
gni_bounds = (Q1_gni - 1.5 * IQR_gni, Q3_gni + 1.5 * IQR_gni)

# Determine outliers
outlier_mask = ((south_asia_latest_clean['hdi'] < hdi_bounds[0]) |
                (south_asia_latest_clean['hdi'] > hdi_bounds[1]) |
                (south_asia_latest_clean['gross_inc_percap'] < gni_bounds[0]) |
                (south_asia_latest_clean['gross_inc_percap'] > gni_bounds[1]))

# Plot
plt.scatter(south_asia_latest_clean[~outlier_mask]['gross_inc_percap'],
            south_asia_latest_clean[~outlier_mask]['hdi'],
            alpha=0.7, label='Normal', color='blue')

plt.scatter(south_asia_latest_clean[outlier_mask]['gross_inc_percap'],
            south_asia_latest_clean[outlier_mask]['hdi'],
            alpha=0.9, label='Outliers', color='red', s=100)

# Annotate outlier countries
for idx, row in south_asia_latest_clean[outlier_mask].iterrows():
    plt.annotate(row['country'], (row['gross_inc_percap'], row['hdi']),
                 xytext=(5, 5), textcoords='offset points')

plt.xlabel('GNI per Capita (USD)')
plt.ylabel('Human Development Index (HDI)')
plt.title('GNI per Capita vs HDI with Outliers Highlighted (South Asia)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('outliers_scatter.png', dpi=300)
plt.show()

# Task 3.3: Discuss outliers
print("\nOutlier Discussion:")
print("1. Countries with exceptionally high GNI but moderate HDI may have:")
print("   - High income inequality")
print("   - Lower investment in health/education")
print("   - Political/social factors limiting development")
print("\n2. Countries with high HDI but lower GNI may have:")
print("   - Effective social policies")
print("   - Good healthcare and education systems")
print("   - Lower cost of living")
```

## TASK 4: Exploring Metric Relationships

```python
# Task 4.1: Select two HDI components
# Using Gender Development Index and Life Expectancy
# Check if gender_development column exists
if 'gender_development' in hdi_south_asia.columns:
    # Task 4.2: Compute Pearson correlation with HDI
    # Filter for latest year and drop missing values
    correlation_data = south_asia_latest.dropna(subset=['hdi', 'gender_development', 'life_expectancy'])

    corr_gender_hdi = correlation_data['gender_development'].corr(correlation_data['hdi'])
    corr_life_hdi = correlation_data['life_expectancy'].corr(correlation_data['hdi'])

    print(f"Correlation between Gender Development Index and HDI: {corr_gender_hdi:.3f}")
    print(f"Correlation between Life Expectancy and HDI: {corr_life_hdi:.3f}")

    # Task 4.3: Create scatter plots with trendlines
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Plot 1: Gender Development vs HDI
    axes[0].scatter(correlation_data['gender_development'], correlation_data['hdi'], alpha=0.7)
    z = np.polyfit(correlation_data['gender_development'], correlation_data['hdi'], 1)
    p = np.poly1d(z)
    axes[0].plot(correlation_data['gender_development'], p(correlation_data['gender_development']),
                "r--", alpha=0.8)
    axes[0].set_xlabel('Gender Development Index')
    axes[0].set_ylabel('HDI')
    axes[0].set_title(f'Gender Development vs HDI (r={corr_gender_hdi:.3f})')
    axes[0].grid(True, alpha=0.3)

    # Plot 2: Life Expectancy vs HDI
    axes[1].scatter(correlation_data['life_expectancy'], correlation_data['hdi'], alpha=0.7)
    z = np.polyfit(correlation_data['life_expectancy'], correlation_data['hdi'], 1)
    p = np.poly1d(z)
    axes[1].plot(correlation_data['life_expectancy'], p(correlation_data['life_expectancy']),
                "r--", alpha=0.8)
    axes[1].set_xlabel('Life Expectancy')
    axes[1].set_ylabel('HDI')
    axes[1].set_title(f'Life Expectancy vs HDI (r={corr_life_hdi:.3f})')
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('metric_relationships.png', dpi=300)
    plt.show()

    # Task 4.4: Discuss relationships
    print("\nMetric Relationships Discussion:")
    print(f"- Life Expectancy shows the strongest relationship with HDI (r={corr_life_hdi:.3f})")
    print("  This indicates health is a crucial component of human development.")
    print(f"- Gender Development Index shows {corr_gender_hdi:.3f} correlation with HDI")
    print("  This suggests gender equality contributes to but doesn't fully determine HDI.")
else:
    print("Gender Development Index column not available in dataset.")
    print("Using education metrics instead...")

    # Alternative: Use mean years of schooling
    correlation_data = south_asia_latest.dropna(subset=['hdi', 'mean_yr_school', 'life_expectancy'])

    corr_edu_hdi = correlation_data['mean_yr_school'].corr(correlation_data['hdi'])
    corr_life_hdi = correlation_data['life_expectancy'].corr(correlation_data['hdi'])

    print(f"\nCorrelation between Mean Years of Schooling and HDI: {corr_edu_hdi:.3f}")
    print(f"Correlation between Life Expectancy and HDI: {corr_life_hdi:.3f}")

    print("\nMetric Relationships Discussion:")
    print(f"- Life Expectancy shows the strongest relationship with HDI (r={corr_life_hdi:.3f})")
    print(f"- Education (Mean Years of Schooling) shows weaker relationship (r={corr_edu_hdi:.3f})")
```

## TASK 5: Gap Analysis

```python
# Task 5.1: Create GNI HDI Gap metric
# Note: Need to normalize both to similar scale first
south_asia_latest['GNI_normalized'] = (south_asia_latest['gross_inc_percap'] -
                                       south_asia_latest['gross_inc_percap'].min()) / \
                                       (south_asia_latest['gross_inc_percap'].max() -
                                        south_asia_latest['gross_inc_percap'].min())

south_asia_latest['HDI_normalized'] = (south_asia_latest['hdi'] -
                                       south_asia_latest['hdi'].min()) / \
                                       (south_asia_latest['hdi'].max() -
                                        south_asia_latest['hdi'].min())

south_asia_latest['GNI_HDI_Gap'] = south_asia_latest['GNI_normalized'] - south_asia_latest['HDI_normalized']

# Task 5.2: Rank countries by GNI HDI Gap
# Descending order (largest positive gaps first)
gap_descending = south_asia_latest.sort_values('GNI_HDI_Gap', ascending=False)
print("\nCountries ranked by GNI-HDI Gap (descending - largest positive gaps):")
print(gap_descending[['country', 'GNI_HDI_Gap', 'gross_inc_percap', 'hdi']].to_string(index=False))

# Ascending order (largest negative gaps first)
gap_ascending = south_asia_latest.sort_values('GNI_HDI_Gap', ascending=True)
print("\nCountries ranked by GNI-HDI Gap (ascending - largest negative gaps):")
print(gap_ascending[['country', 'GNI_HDI_Gap', 'gross_inc_percap', 'hdi']].to_string(index=False))

# Task 5.3: Plot top 3 positive gaps and top 3 negative gaps
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Positive gaps (GNI > HDI)
top_positive = gap_descending.head(3)
axes[0].barh(top_positive['country'], top_positive['GNI_HDI_Gap'], color='red')
axes[0].set_xlabel('GNI-HDI Gap (Positive = GNI > HDI)')
axes[0].set_title('Top 3 Positive GNI-HDI Gaps')
axes[0].invert_yaxis()

# Negative gaps (HDI > GNI)
top_negative = gap_ascending.head(3)
axes[1].barh(top_negative['country'], top_negative['GNI_HDI_Gap'], color='green')
axes[1].set_xlabel('GNI-HDI Gap (Negative = HDI > GNI)')
axes[1].set_title('Top 3 Negative GNI-HDI Gaps')
axes[1].invert_yaxis()

plt.tight_layout()
plt.savefig('gni_hdi_gaps.png', dpi=300)
plt.show()

# Task 5.4: Discuss implications
print("\nImplications of GNI-HDI Gap:")
print("\nPositive Gap (GNI > HDI):")
print("- High income but lower human development")
print("- Possible reasons: Income inequality, poor social services,")
print("  inefficient use of economic resources for development")
print("- Example countries:", list(top_positive['country']))

print("\nNegative Gap (HDI > GNI):")
print("- Better human development relative to income level")
print("- Possible reasons: Effective social policies, good governance,")
print("  efficient use of limited resources for development")
print("- Example countries:", list(top_negative['country']))

print("\nPolicy Implications:")
print("1. Countries with positive gaps should invest more in health/education")
print("2. Countries with negative gaps show efficient development models")
print("3. Gap analysis helps identify development inefficiencies")
```

```python
print("\nTask 3: Top and Bottom Performers")
print("=" * 50)

# Function to get top and bottom performers for a region
def get_top_bottom_performers(region_df, n=3):
    # Use average HDI across 2020-2022 for each country
    country_avg_hdi = region_df.groupby('country')['hdi'].mean().sort_values(ascending=False)

    top_countries = country_avg_hdi.head(n)
    bottom_countries = country_avg_hdi.tail(n)

    return top_countries, bottom_countries

# Get top and bottom for South Asia
south_asia_top, south_asia_bottom = get_top_bottom_performers(hdi_south_asia_2020_2022)
print("South Asia - Top 3 performers:")
print(south_asia_top)
print("\nSouth Asia - Bottom 3 performers:")
print(south_asia_bottom)

# Get top and bottom for Middle East
middle_east_top, middle_east_bottom = get_top_bottom_performers(hdi_middle_east_2020_2022)
print("\nMiddle East - Top 3 performers:")
print(middle_east_top)
print("\nMiddle East - Bottom 3 performers:")
print(middle_east_bottom)

# Create bar chart comparing top and bottom performers
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# South Asia Top 3
axes[0, 0].bar(south_asia_top.index, south_asia_top.values, color='blue')
axes[0, 0].set_title('South Asia - Top 3 HDI Performers')
axes[0, 0].set_ylabel('Average HDI (2020-2022)')
axes[0, 0].tick_params(axis='x', rotation=45)

# South Asia Bottom 3
axes[0, 1].bar(south_asia_bottom.index, south_asia_bottom.values, color='red')
axes[0, 1].set_title('South Asia - Bottom 3 HDI Performers')
axes[0, 1].set_ylabel('Average HDI (2020-2022)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Middle East Top 3
axes[1, 0].bar(middle_east_top.index, middle_east_top.values, color='green')
axes[1, 0].set_title('Middle East - Top 3 HDI Performers')
axes[1, 0].set_ylabel('Average HDI (2020-2022)')
axes[1, 0].tick_params(axis='x', rotation=45)

# Middle East Bottom 3
axes[1, 1].bar(middle_east_bottom.index, middle_east_bottom.values, color='orange')
axes[1, 1].set_title('Middle East - Bottom 3 HDI Performers')
axes[1, 1].set_ylabel('Average HDI (2020-2022)')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('top_bottom_performers.png', dpi=300)
plt.show()
```

## TASK 4: Metric Comparisons

```python
print("\nTask 4: Metric Comparisons")
print("=" * 50)

# Ensure numeric data types for comparison metrics
metrics_to_compare = ['gender_development', 'life_expectancy', 'gross_inc_percap']

# Calculate average for each metric by region
region_comparison = {}

for metric in metrics_to_compare:
    if metric in hdi_south_asia_2020_2022.columns:
        # Convert to numeric
        hdi_south_asia_2020_2022[metric] = pd.to_numeric(hdi_south_asia_2020_2022[metric], errors='coerce')
        hdi_middle_east_2020_2022[metric] = pd.to_numeric(hdi_middle_east_2020_2022[metric], errors='coerce')

        south_asia_avg = hdi_south_asia_2020_2022[metric].mean()
        middle_east_avg = hdi_middle_east_2020_2022[metric].mean()

        region_comparison[metric] = {
            'South Asia': south_asia_avg,
            'Middle East': middle_east_avg,
            'Difference': middle_east_avg - south_asia_avg
        }

# Display comparison
print("Metric Comparison (Average Values):")
print(f"{'Metric':25} {'South Asia':>12} {'Middle East':>12} {'Difference':>12}")
print("-" * 65)

for metric, values in region_comparison.items():
    metric_name = metric.replace('_', ' ').title()
    print(f"{metric_name:25} {values['South Asia']:12.2f} {values['Middle East']:12.2f} {values['Difference']:12.2f}")

# Identify metric with greatest disparity
metric_with_max_diff = max(region_comparison.items(), key=lambda x: abs(x[1]['Difference']))
print(f"\nMetric with greatest disparity: {metric_with_max_diff[0].replace('_', ' ').title()}")
print(f"Disparity: {abs(metric_with_max_diff[1]['Difference']):.2f}")

# Create grouped bar chart
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

metric_names = {
    'gender_development': 'Gender Development Index',
    'life_expectancy': 'Life Expectancy',
    'gross_inc_percap': 'GNI per Capita'
}

for idx, (metric, values) in enumerate(region_comparison.items()):
    regions = ['South Asia', 'Middle East']
    avg_values = [values['South Asia'], values['Middle East']]

    bars = axes[idx].bar(regions, avg_values, color=['blue', 'green'])
    axes[idx].set_title(metric_names[metric])
    axes[idx].set_ylabel('Average Value')

    # Add value labels
    for bar in bars:
        height = bar.get_height()
        axes[idx].text(bar.get_x() + bar.get_width()/2., height + height*0.01,
                      f'{height:.2f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('metric_comparison_grouped_bars.png', dpi=300)
plt.show()
```

## TASK 5: HDI Disparity

```python
print("\nTask 5: HDI Disparity")
print("=" * 50)

# Function to calculate HDI statistics for a region
def calculate_hdi_disparity(region_df, region_name):
    hdi_values = region_df['hdi'].dropna()

    hdi_range = hdi_values.max() - hdi_values.min()
    hdi_mean = hdi_values.mean()
    hdi_std = hdi_values.std()
    coefficient_variation = (hdi_std / hdi_mean) * 100 if hdi_mean != 0 else 0

    return {
        'Region': region_name,
        'Range': hdi_range,
        'Mean': hdi_mean,
        'Std': hdi_std,
        'CV (%)': coefficient_variation
    }

# Calculate for both regions
south_asia_stats = calculate_hdi_disparity(hdi_south_asia_2020_2022, 'South Asia')
middle_east_stats = calculate_hdi_disparity(hdi_middle_east_2020_2022, 'Middle East')

# Display results
print("HDI Disparity Analysis:")
print(f"{'Statistic':15} {'South Asia':>12} {'Middle East':>12}")
print("-" * 45)

for key in ['Mean', 'Range', 'Std', 'CV (%)']:
    print(f"{key:15} {south_asia_stats[key]:>12.3f} {middle_east_stats[key]:>12.3f}")

# Identify which region exhibits more variation
if south_asia_stats['CV (%)'] > middle_east_stats['CV (%)']:
    print(f"\nSouth Asia exhibits more variation in HDI (CV: {south_asia_stats['CV (%)']:.1f}% vs {middle_east_stats['CV (%)']:.1f}%)")
else:
    print(f"\nMiddle East exhibits more variation in HDI (CV: {middle_east_stats['CV (%)']:.1f}% vs {south_asia_stats['CV (%)']:.1f}%)")

# Create visualization of HDI distributions
plt.figure(figsize=(10, 6))

# Box plot for HDI distributions
data = [hdi_south_asia_2020_2022['hdi'].dropna(),
        hdi_middle_east_2020_2022['hdi'].dropna()]

box = plt.boxplot(data, labels=['South Asia', 'Middle East'], patch_artist=True)
plt.ylabel('HDI Value')
plt.title('HDI Distribution Comparison: South Asia vs Middle East')
plt.grid(True, alpha=0.3)

# Color boxes
colors = ['lightblue', 'lightgreen']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.tight_layout()
plt.savefig('hdi_distribution_comparison.png', dpi=300)
plt.show()
```

## TASK 6: Correlation Analysis

```python
print("\nTask 6: Correlation Analysis")
print("=" * 50)

# Function to calculate correlations for a region
def calculate_correlations(region_df, region_name):
    correlations = {}

    # Check if gender_development exists
    if 'gender_development' in region_df.columns:
        # Convert to numeric
        region_df['gender_development'] = pd.to_numeric(region_df['gender_development'], errors='coerce')
        corr_gender_hdi = region_df[['gender_development', 'hdi']].dropna().corr().iloc[0, 1]
        correlations['Gender Development vs HDI'] = corr_gender_hdi

    # Life Expectancy correlation
    region_df['life_expectancy'] = pd.to_numeric(region_df['life_expectancy'], errors='coerce')
    corr_life_hdi = region_df[['life_expectancy', 'hdi']].dropna().corr().iloc[0, 1]
    correlations['Life Expectancy vs HDI'] = corr_life_hdi

    return correlations

# Calculate correlations for both regions
south_asia_corr = calculate_correlations(hdi_south_asia_2020_2022, 'South Asia')
middle_east_corr = calculate_correlations(hdi_middle_east_2020_2022, 'Middle East')

print("Correlation Analysis:")
print(f"{'Correlation':30} {'South Asia':>12} {'Middle East':>12}")
print("-" * 60)

for metric in ['Gender Development vs HDI', 'Life Expectancy vs HDI']:
    if metric in south_asia_corr:
        print(f"{metric:30} {south_asia_corr[metric]:>12.3f} {middle_east_corr.get(metric, 'N/A'):>12}")

# Create scatter plots with trendlines
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# South Asia correlations
if 'Gender Development vs HDI' in south_asia_corr:
    # Gender Development vs HDI for South Asia
    sa_gender_data = hdi_south_asia_2020_2022[['gender_development', 'hdi']].dropna()
    axes[0, 0].scatter(sa_gender_data['gender_development'], sa_gender_data['hdi'], alpha=0.6)
    z = np.polyfit(sa_gender_data['gender_development'], sa_gender_data['hdi'], 1)
    p = np.poly1d(z)
    axes[0, 0].plot(sa_gender_data['gender_development'], p(sa_gender_data['gender_development']),
                   'r--', alpha=0.8)
    axes[0, 0].set_xlabel('Gender Development Index')
    axes[0, 0].set_ylabel('HDI')
    axes[0, 0].set_title(f'South Asia: Gender Dev vs HDI (r={south_asia_corr["Gender Development vs HDI"]:.3f})')
    axes[0, 0].grid(True, alpha=0.3)

# Life Expectancy vs HDI for South Asia
sa_life_data = hdi_south_asia_2020_2022[['life_expectancy', 'hdi']].dropna()
axes[0, 1].scatter(sa_life_data['life_expectancy'], sa_life_data['hdi'], alpha=0.6)
z = np.polyfit(sa_life_data['life_expectancy'], sa_life_data['hdi'], 1)
p = np.poly1d(z)
axes[0, 1].plot(sa_life_data['life_expectancy'], p(sa_life_data['life_expectancy']),
               'r--', alpha=0.8)
axes[0, 1].set_xlabel('Life Expectancy')
axes[0, 1].set_ylabel('HDI')
axes[0, 1].set_title(f'South Asia: Life Expectancy vs HDI (r={south_asia_corr["Life Expectancy vs HDI"]:.3f})')
axes[0, 1].grid(True, alpha=0.3)

# Middle East correlations
if 'Gender Development vs HDI' in middle_east_corr:
    # Gender Development vs HDI for Middle East
    me_gender_data = hdi_middle_east_2020_2022[['gender_development', 'hdi']].dropna()
    axes[1, 0].scatter(me_gender_data['gender_development'], me_gender_data['hdi'], alpha=0.6)
    z = np.polyfit(me_gender_data['gender_development'], me_gender_data['hdi'], 1)
    p = np.poly1d(z)
    axes[1, 0].plot(me_gender_data['gender_development'], p(me_gender_data['gender_development']),
                   'r--', alpha=0.8)
    axes[1, 0].set_xlabel('Gender Development Index')
    axes[1, 0].set_ylabel('HDI')
    axes[1, 0].set_title(f'Middle East: Gender Dev vs HDI (r={middle_east_corr["Gender Development vs HDI"]:.3f})')
    axes[1, 0].grid(True, alpha=0.3)

# Life Expectancy vs HDI for Middle East
me_life_data = hdi_middle_east_2020_2022[['life_expectancy', 'hdi']].dropna()
axes[1, 1].scatter(me_life_data['life_expectancy'], me_life_data['hdi'], alpha=0.6)
z = np.polyfit(me_life_data['life_expectancy'], me_life_data['hdi'], 1)
p = np.poly1d(z)
axes[1, 1].plot(me_life_data['life_expectancy'], p(me_life_data['life_expectancy']),
               'r--', alpha=0.8)
axes[1, 1].set_xlabel('Life Expectancy')
axes[1, 1].set_ylabel('HDI')
axes[1, 1].set_title(f'Middle East: Life Expectancy vs HDI (r={middle_east_corr["Life Expectancy vs HDI"]:.3f})')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('correlation_analysis.png', dpi=300)
plt.show()

# Interpretation
print("\nInterpretation of Correlations:")
print("1. Strength: Correlation coefficients closer to 1 or -1 indicate stronger relationships.")
print("2. Direction: Positive correlations mean as one variable increases, the other also increases.")
print("3. Regional differences: Compare South Asia vs Middle East correlation strengths.")
```
